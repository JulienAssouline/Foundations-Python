### How many total results are there? How do you change the URL to get more?

There are a few ways to change the URL to: there is one very easy way and one a-little-difficult way.

```python
url_women_tops = "http://api.hm.com/v2/us/en/products/display?categories=sale%2Fladies%2Ftops&concealCategories=true&pageSize=500&page=1&deviceType=DESKTOP"
response_women = requests.get(url_women_tops)
data_women = response_women.json()
len(data_women["displayArticles"])





```

### Download all on-sale women's tops and save them to a CSV.

I want this CSV to include

- Name
- Product code/number
- Old price
- On-sale price
- Item's URL

```python
womens_prices = []
for i in data_women["displayArticles"]:
    dictionary = {}
    priceInfo = i["priceInfo"]
    if priceInfo:
        print(priceInfo["formattedPrice"])
        print(priceInfo["formattedOldPrice"])
        dictionary["new price"] = priceInfo["formattedPrice"]
        dictionary["old price"] = priceInfo["formattedOldPrice"]
    productNumber = i["productNumber"]
    if productNumber:
        print(productNumber)
        dictionary["product number"] = productNumber
    name = i["name"]
    if name:
        print(name)
        dictionary["name"] = name
    webUrl = i["webUrl"]
    if webUrl:
        print(webUrl)
        dictionary["url"] = webUrl
    womens_prices.append(dictionary)
```

```python
womens_prices
```

## Analyzing on-sale items

### Read in the on-sale CSV and make sure the data looks right

```python
import pandas as pd

df = pd.DataFrame(womens_prices)


df.to_csv("H&M womens tops sale.csv", index = False)

df = pd.read_csv("H&M womens tops sale.csv")
df.head()

```

### Convert the prices to numbers, if you haven't already

```python
df["new price"] = df["new price"].str.replace("$", "")
df["old price"] = df["old price"].str.replace("$", "")
df.head()
#df['returns'].str[:-1].astype(float))
```

```python
df["new price"] = df["new price"].astype(float)
df["old price"] = df["old price"].astype(float)

df.head()
df.dtypes
```

### What is the average on-sale price of a women's top? Make a histogram, too, please.

```python
%matplotlib inline

df["new price"].mean()
```

```python
df["new price"].hist()
```

### What is the average discount between full price and on sale? Histogram too, please.

```python
df["discount"] = df["old price"] - df["new price"]
df["discount"].mean()
```

```python
df["discount"].hist()
```

### What is the average discount as a percentage of the full price? Histogram too, please.

```python
df["proportion"] = (df["new price"] / df["old price"]) * 100
df["proportion"].mean()
```

```python
df["proportion"].hist()
```

## Shopping the sales

### How many items are over 50% off?


```python
df["over_50?"] = df["proportion"] > 50
df["over_50?"].value_counts()
```

### Display all items above 50% off.
